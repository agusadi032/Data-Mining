```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv('diabetes.csv')
dataset.keys()
```

```python
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
X = np.asarray(dataset)
print(X)
```

Menampilkan data Insulin dan Glocose dalam bentuk scatter plot

```python
plt.scatter(X[:,1], X[:,4], label='True Position')
plt.xlabel("Insulin")
plt.ylabel("Glucose")
plt.title("Grafik Data Insulin dan Glukosa")
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
print(kmeans.cluster_centers_)
```

Hasil dari klusterisasi data insulin dan glucose

```python
plt.scatter(X[:,1], X[:,4], c=kmeans.labels_, cmap= 'rainbow')
plt.xlabel("Insulin")
plt.ylabel("Glucose")
plt.title("Grafk Hasil Klasterisasi data Insulin dan Glucose")
plt.show()
```

Menampilkan hasil klusterisasi dengan centeroid dari data Insulin dan Glucose

```python
plt.scatter(X[:,1], X[:,4], c=kmeans.labels_, cmap= 'rainbow')
plt.scatter(kmeans.cluster_centers_[:,1], kmeans.cluster_centers_[:,4], color= 'black')
plt.xlabel("Insulin")
plt.ylabel("Glucose")
plt.title("Grafk Hasil Klasterisasi data Centeroid Insulin dan Glucose")
plt.show()
```

Menampilkan data dari Pregnancise dan Age

```python
plt.scatter(X[:,0], X[:,7], label='True Position')
plt.xlabel("Pregnancise")
plt.ylabel("Age")
plt.title("Grafik Data Preganancise dan Age")
plt.show()
```

Hasil Klasterisasi dari data Pregnancise dan Age 

```python
plt.scatter(X[:,0], X[:,7], c=kmeans.labels_, cmap= 'rainbow')
plt.xlabel("Pregnancise")
plt.ylabel("Age")
plt.title("Grafk Hasil Klasterisasi data Pregnancise dan Age")
plt.show()
```

Menampilkan Hasil Klusterisasi dengan Centeroid data dari Pregnancise dan Age

```python
plt.scatter(X[:,0], X[:,7], c=kmeans.labels_, cmap= 'rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,7], color= 'black')
plt.xlabel("Pregnancise")
plt.ylabel("Age")
plt.title("Grafk Hasil Klasterisasi data Centeroid Pregnancise dan Age")
plt.show()
```
